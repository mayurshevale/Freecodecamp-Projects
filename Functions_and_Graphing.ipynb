## **Function and a table of (x,y) values**

```python
x = 5
y = 4*x + 3 # Function
print(x,",",y)
print("x \t y")

for x in range(11):
  y = 4*x + 3
  print(x,"\t",y)

```

**Defining a Python Function**

```python
def f(x):
  y = 4*x + 3
  return y
print(5,",",f(5))

# Next a loop
for x in range(11):
  print(x,"\t",f(x))
```

**Using a Exception in the code**

```python
def f(x):
    try:
        if not isinstance(x, int):
            raise ValueError("Input should be int")
        y = 4 * x + 3
        return y
    except ValueError as ve:
        print("Caught a ValueError: {}".format(ve))

print(5, ",", f(5))

# Next a loop
for x in range(11):
          print(x, "\t", f(x))
```

**Display axis line**

```python
import matplotlib.pyplot as plt
xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # Window size
plt.plot([xmin,xmax],[0,0],"b") # Blue x axis
plt.plot([0,0],[ymin,ymax],"b") # Blue y axis
plt.show()
```

 **Use an Array as input**

```python
import matplotlib.pyplot as plt
import numpy as np
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2 * (xmax - xmin)
x = np.linspace(xmin,xmax,points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # Window size
plt.plot([xmin,xmax],[0,0],"b") # Blue x axis
plt.plot([0,0],[ymin,ymax],"b") # Blue y axis

y = 2*x + 1
plt.plot(x,y,"r")
  
plt.show()

```

**More Ways to customize the graph**

```python
import matplotlib.pyplot as plt
import numpy as np
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2 * (xmax - xmin)
x = np.linspace(xmin,xmax,points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # Window size
plt.plot([xmin,xmax],[0,0],"b") # Blue x axis
plt.plot([0,0],[ymin,ymax],"b") # Blue y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Sample graph")
ax.grid(True)

ax.set_xticks(np.arange(xmin,xmax,2))
ax.set_yticks(np.arange(ymin,ymax,2))

y = 2*x + 1
plt.plot(x,y,label="y = 2*x + 1")
plt.plot([4],[6],'ro',label='point')
plt.plot(x,3*x,'k',label='steeper line')

plt.legend(loc='upper left')  
plt.show()

```
